```python
import numpy as np
np.random.seed(254)
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0,np.pi,600)
inp = np.sin(t)
plt.plot(t,inp)
```

```python
trigger = False
hi_thrshld = np.sin(np.pi/3)
low_thrshld = 0.5

hi_val = 1.5
low_val = -0.5
op = []
for x in inp:
    if not trigger:
        if x > hi_thrshld:
            trigger = True
            y = hi_val
        else:
            y = low_val
    if trigger:
        if x < low_thrshld:
            trigger = False
            y = low_val
        else:
            y = hi_val
    op.append(y)
        
```

```python
plt.plot(op)
plt.plot(inp)
plt.xticks(np.linspace(0,600,7),np.linspace(0,6,7));
plt.grid(True)
```
